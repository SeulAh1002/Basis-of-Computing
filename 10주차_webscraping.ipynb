```
import requests

target_url=input('URL 입력 : ')

if (target_url[0:7] != 'httpL//') :
  target_url= 'http://' + target_url

rs= requests.get(target_url)

if (rs.status_code ==200):
  print(rs.cookies)
else:
  print('Fail: Check the Status Code Below')
  print(rs.status_code)
```

```
from urllib.request import urlopen
from bs4 import BeautifulSoup

page=urlopen('https://finance.naver.com/marketindex/')
soup=BeautifulSoup(page, 'html.parser')

every_region= soup.find_all('h3', 'h_lst')
every_rate= soup.find_all('span', 'value')

i=0
num_region= len(every_region)

while (i < num_region) : 
  region = every_region[i].string
  rate= every_rate[i].string
  print(region, '->', rate)
  i=i+1
```

```
from urllib.request import urlopen
from bs4 import BeautifulSoup

page=urlopen('https://finance.naver.com/marketindex/exchangeList.nhn')
soup=BeautifulSoup(page, 'html.parser')

every_data= soup.find_all('td')

i=0
num_data= len(every_data)

while (i < num_data) : 
  if (i % 7 == 0) :  
    region = every_data[i].string
    region= region.replace('\n','')
    region= region.replace('\t','')
    print(region, end=' ')
  else:
    rate = every_data[i].string
    print(rate, end= ' ')
  i=i+1
  if (i%7 ==0):
    print()
```
